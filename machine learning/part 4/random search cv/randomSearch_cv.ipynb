```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
# As, we predict species, so lets check the values in that
df['species'].unique()
```

```python
# now split the X and y
X = df.drop('species', axis=1)
y = df['species']
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# knn model

```python
# import library
from sklearn.neighbors import KNeighborsClassifier
# set parameters
model_knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# train the model
model_knn.fit(X_train, y_train)
```

```python
# see the model score/accuracy
model_knn.score(X_test, y_test)
```

Note: If we change the value of n_neighbors like, 9, 10, 13 etc, then the  model score also changed. Thats because we changed the parameters of model

# svm model

```python
from sklearn.svm import SVC
# set parameters
model_svm = SVC(C= 30, kernel='rbf', gamma='auto')
```

```python
# train the model
model_svm.fit(X_train, y_train)
```

```python
# see the model score/accuracy
model_svm.score(X_test, y_test)
```

Note: here also if we change the value of kernel='linear' and C like 10, 5, 12 etc, then the model score also changed. Thats because we changed the parameters of model

# Random search cv

```python
# lets perform the grid search cv on svm
from sklearn.svm import SVC

model_svm = SVC(gamma='auto')
# train the model
model_svm.fit(X_train, y_train)
```

```python
# see the model score/accuracy
model_svm.score(X_test, y_test)
```

**now, perform the random search cv on svm**

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# set the parametrs on which we do cross validations multiple times
cv_classifier = RandomizedSearchCV((model_svm), {
    'C' : [1,10,20,30],
    'kernel' : ['rbf', 'linear']
}, cv=5, return_train_score=False, n_iter=4)
```

```python
# fit the X and y
cv_classifier.fit(X, y)
```

```python
# to check the results
cv_classifier.cv_results_
```

```python
# Its not understandable, so we convert it into dataFrame
```

```python
results = pd.DataFrame(cv_classifier.cv_results_)
```

```python
results.head()
```

here, we need to see the parametes and output score, so we extract them from whole results

```python
# extrat parametes that we applied and output score
results[['param_C', 'param_kernel', 'mean_test_score']]
```

here, we find the 4 scores, because we put n_iter=4. So out of all results, it show us 4 results

Note: We can also apply the random_search_cv on knn also, by creating the parameeters that we want to put as cross validations

Note: We use random_search_cv, when we have hundreds or thousands of combinations of cross validations. And we select the some cross validations by giving the n_iter value
