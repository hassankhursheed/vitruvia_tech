```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
X, y = make_blobs(n_samples=500, centers=3, n_features=5, cluster_std=1.5, random_state=42)
```

```python
X
```

```python
# convert the X in StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# you can check the X scaled values
X_scaled
```

# PCA

```python
# call PCA and set parameters
pca = PCA(n_components=2)
```

```python
# fit the X scaled value in PCA
X_pca = pca.fit_transform(X_scaled)
```

```python
# lets see the X pca that we made as n_components=2
X_pca
```

```python
# now convert the pca values in data frame
df_pca = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
df_pca.head()
```

```python
# extract labels from pca
df_pca['label'] = y
```

```python
# now plot it
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='label', palette='Set2')
```
