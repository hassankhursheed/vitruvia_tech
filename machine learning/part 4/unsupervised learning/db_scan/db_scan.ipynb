```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
```

```python
X, y_true = make_moons(n_samples=500, noise=0.05, random_state=42)
```

```python
# X

# convert the X in DataFrame
df = pd.DataFrame(X, columns=['Feature_1', 'Feature_2'])
df.head()
```

```python
# standard Scaled the df of X
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

# k means algorithm

```python
# now set the kmeans parameters value
kmeans = KMeans(n_clusters=2, random_state=42)
```

```python
# set the kmeans labels and fit it
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
# put the kmeans_labels in df
df['kmeans_cluster'] = kmeans_labels
```

```python
# now plot the kmeans_labels
sns.scatterplot(x=df['Feature_1'], y=df['Feature_2'], hue=df['kmeans_cluster'], palette='tab10')
```

Note: here the problem is that is doesn't make the cluster of grouped data. rather it make cluster by following a center shape. Here, its proved that k means is centroid_based algorithm

# DB scan algorithm

```python
# now set the DB scan parameters value
dbscan = DBSCAN(eps=0.3, min_samples=5)
```

```python
# set the dbscan labels and fit it
dbscan_labels = dbscan.fit_predict(X_scaled)
```

```python
# put the dbscan_labels in df
df['dbscan_cluster'] = dbscan_labels
```

```python
# now plot the ddbscan_labels
sns.scatterplot(x=df['Feature_1'], y=df['Feature_2'], hue=df['dbscan_cluster'], palette='tab10')
```

Note: here, it makes clusters by following the density of points and make clusters of each grouped points separately. Here, its proved that DB scan is density_based algorithm
